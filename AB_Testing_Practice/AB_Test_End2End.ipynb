# 1. What is AB Testing and when to use it?
The goal of A/B Testing is to show a new variation of a product to an experiemental group, and an existing version of the product to a control group. The goal is to track the metric during the test period and determine if there is a difference in performance/efficacy of the product - and what kind of difference is it.

Benefits of A/B Testing:

    - Quickly learn what works and what doesn't
    - Direct feedback from customers
    - Results are unbiased since testers are unaware
    
Drawbacks of A/B Testing:

    - Presenting different products/featurs/etc may be dangerous and casuse Change Aversion
    - Requires significant resources
    - Can lead to incorrect solutions if not conducted properly

# 2. Questions to ask before an A/B test

    - Can we find the answer using exploratory/historical data analysis?

    - Can we assure truly randomized control and experiemental groups such that both samples are an unbiasedand 
        true representation of the true user population
    - Can we ensure teh intefrity of the treatment vs control eggects during the entire duration of the test?

# 3. Choosing Primary Metric for the A/B Test
It is important to chose the metric that will be used to measure the performance of the product. It will also be used to identigy whether there is a statistically significant difference between these two groups.

The choice of the *success metric* depends on the underlying hypotheis that is being tested with the A/B Test. We need to tie up the primary metric to the direcyt and higher level goals of the product. One way to test the accuracy of a metric you have chosen for your A/B Test could be to go back to the exact problem you want to solve.

*Metric Validity Question: If this chosen metric were to increase significantly while everything else stays constant, would we achieve our goal address the problem?*

Though you need to have a single primary metric for the A/B Test, we still need to keep an eye on the remaining metrics to make sure all the metrics are showing change and not only the target one. Having *multiple* metrics in the test will lead to false positives since you will identify many signigicant differences while there is no effect.

# 4. Common A/B Test Metrics
a. Click Through Rate: The percent of people who view the page (impressions) and then actually click on it

c. Conversion Rate: The proportion of sessions ending up with a transaction

# 6. Designing the A/B Test
   ### Step 1: Statistical Hypothesis
        a. Null:
            - probability of control = probability of experiment
            - population mean of control = population mean of experiment
        b. Alternate:
            - probability of control != probability of experiment
            - population mean of control != population mean of experiment
     
     
   ### Step 2: Power Analysis 
        Before running an A/B Test we need to determine the sample size of the control and
         experiemental groups and for how long we need to run the test. 
         This is known as the Power Analysis and it includes 3 steps:
             
         a. Determining the power of the test
             - The probability of correctly rejecting the null hypothesis when it is false.
             - Power = probability of not makeing a type 2 error
             - Common to pick 80% as the power of the test - 20% type 2 error which means we are fine with 
                 not detecting (failing to reject the null) a tratment effect while there is an effect
                 
         b. Determining the signigicance level of the test
             - Significance level = the probability of Type 1 error is the likelihood of rejecting the null,
                 hence detecting the treatment effect, while the null is true and there is no stat sig. impact.
                 (False pos rate)
             - Significance value of 5% indicates a 5% risk of concluding that there exists a stat sig. 
                 difference btwn the experimental and the control when there is no actual difference
                     
         c. Determining a Minimum Detectable Effect 
             - MDE = what is the substantive to the stat sig. that the business wants to see as a 
                 min impact of the new product?
             - What is the amount of change we aim to observe in the new product compared to 
                 the existing one to recommend to the business that the feature should be launched
             - MDE is a proxy that related to the smallest effect that would matter in practice for the business 
                 (usually set by stakeholders)


   ### Step 3: Calculating Minimum Sample Size
        a. Sample size must be determined using the:
            - power of the test
            - significance level
            - MDE
            - variances of the 2 normally distributed samples of equal size
            
        b. Case 1: Sample Size (N) Calculation with Binary Metric
            - CTR: user can click (success) or not (failure) and if the users' responses to the product
                can be defined as independent we can consider this *Bernoulli Trials* 
            - Bernoulli Trials: where the click event (success) occurs with probability p_con, p_exp (control/exp)
                - q_con = 1 - p_con
                - q_exp = 1 - p_con
            - The sample size needed to compare these two Binomial Proportions using a two sided test with prespecified
                sig level, power level, MID.
            - We need to use A/A testing (AB test except assigning the same treatment to both groups) 
                to obtain estimates.

         
         
        c. Case 2: Sample Size (N) Calculation with Continuous Metric
            - Primary performance metric that is in the form of an average (mean order amount) where we intend
                to compare the means of the contol & exp groups, then use the CLT and state the mean sampling
                distribution of both groups follow normal distribution
            - The same size needed to compare the means of 2 normally distributed samples using a 2 sided test with
                prespecified sig level, power, and MID
            - We can run an A/A test to obtain the sample variances _con, and _exp.

    
    
   ### Step 4: Determining A/B test duration
        a. Baseline duration time = N / # visitors per day
        b. Take into account business aspects (ex. Covid)

## 8a. 2-Sample T-Test
If you want to test whether there is a statistically significant difference between the control and experiemental groups' metrics
    that are in the form of averages, metric follows student-t distribution and N < 30

```python
import numpy as np
from scipy.stats import t

np.random.seed(18)

N_con = 20
df_con = N_con - 1 # degrees of freedom of Control
N_exp = 20
df_exp = N_exp - 1 # degrees of fredom of Experiment

# Significance Level
alpha = 0.05

# Data of control group with t-distribution
X_con = np.random.standard_t(df_con, N_con)
# Data of experimental group with t-distribution
X_exp = np.random.standard_t(df_exp, N_exp)

# Mean of Control
mu_con = np.mean(X_con)
# Mean of Experimental
mu_exp = np.mean(X_exp)

# Variance of Control
sigma_sqr_con = np.var(X_con)
# Variance of Experimental
sigma_sqr_exp = np.var(X_exp)

# Pooled Variance
pooled_var_t_test = ((N_con - 1) * sigma_sqr_con + (N_exp - 1) * sigma_sqr_exp) / (N_con + N_exp - 2)*(1/N_con + 1/N_exp)

# Standard Error
se = np.sqrt(pooled_var_t_test)

# Test Statistics
T = (mu_con - mu_exp) / se

# Critical value for 2 sided sample t test
t_crit = t.ppf(1-alpha/2, N_con + N_exp - 2)

# P value of the 2 sided t testm using t-distribution and its symmetric property
p_value = t.sf(T, N_con + N_exp - 2) * 2

# Margin of Error
margin_error  = t_crit * se

# Confidence Interval
ci = [(mu_con - mu_exp) - margin_error, (mu_con - mu_exp) + margin_error]

print("T-score: ", T)
print("T-critical: ", t_crit)
print("P_value: ", p_value)
print("Confidence Interval of 2 sample Z-test: ", np.round(ci,2))
```

## 8b. 2-Sample Z-Test
If you want to test whether there is a statistically significant difference between control and experiemental groups' metrics that are in the form of averages or proportions, metric follows normal distribution, N > 30 such that you can use CLT.

### Case 1: Z-Test for Comparing Proportions
Metrics that are in the form of proportions (CTR) and if the click event occurs independently.
    - Each click can be described by a random var that can only be 1 (success) or 0 (failure) 
        and that follows a Bernoulli distribution  (click: success and no click: failure).

```python
import numpy as np
from scipy.stats import norm

# Control clicks
X_con = 1242 
# Control impressions
N_con = 9886
# Experiement clicks
X_exp = 974
# Experiement impressions
N_exp = 10072

# Significance level
alpha = 0.05

# Probability of interaction for Control and Experiment
p_con_hat = X_con / N_con
p_exp_hat = X_exp / N_exp

# Calculate the pooled probablity for success and pooled variance to test diff in probabilities
p_pooled_hat = (X_con + X_exp) / (N_con + N_exp)
pooled_variance = p_pooled_hat * (1 - p_pooled_hat) * (1/N_con + 1/N_exp)

# Standard error
SE = np.sqrt(pooled_variance)

# Test statistics
test_stat = (p_con_hat - p_exp_hat) / SE
# Critical value using the standard normal distribution
z_crit = norm.ppf(1 - alpha/2)

# Margin of error
m = SE * z_crit
# Two sided test and using symmetry property of Normal distribution so we multiply by 2
p_value = norm.sf(test_stat) * 2

# Confidence Internval
CI = [(p_con_hat - p_exp_hat) - SE * z_crit, (p_con_hat - p_exp_hat) + SE * z_crit]

if np.abs(test_stat) >= z_crit:
    print("Reject the null")
    print(p_value)
    
print("Test statistic stat: ", test_stat)
print("Z-critical: ", z_crit)
print("P-value: ", p_value)
print("Confidence Interval of 2 sample Z-test for proportions", np.round(CI,2))

import matplotlib.pyplot as plt
z = np.arange(-3,3,  0.1)
plt.plot(z, norm.pdf(z), label = 'Standard Normal Distribution',color = 'purple',linewidth = 2.5)
plt.fill_between(z[z>z_crit], norm.pdf(z[z>z_crit]), label = 'Right Rejection Region',color ='y' )
plt.fill_between(z[z<(-1)*z_crit], norm.pdf(z[z<(-1)*z_crit]), label = 'Left Rejection Region',color ='y' )
plt.title("Two Sample Z-test rejection region")
plt.legend()
plt.show()
```

### Case 2: Z-Test for Comparing Means
Metrics that are in the form of averages (CTR).

    - Sampling distribution of means of Control and Experiment group follows Normal distribution

```python
import numpy as np
from scipy.stats import norm

N_con = 60
N_exp = 60

# Significance level
alpha = 0.05

X_con = np.random.randint(100, size = N_con)
X_exp = np.random.randint(100, size = N_exp)

# Calculating means of control and experiment
mu_con = np.mean(X_con)
mu_exp = np.mean(X_exp)

# Calculate variances
variance_con = np.var(X_con)
variance_exp = np.var(X_exp)

# Pooled Variance
pooled_variance = np.sqrt(variance_con/N_con + variance_exp/N_exp)

# Test Statistics
T = (mu_con - mu_exp) / np.sqrt(variance_con/N_con + variance_exp/N_exp)

# Two sided test and using symmetry property of Normal distribution so we multiply by 2
p_value = norm.sf(T)*2

# Z critical value
Z_crit = norm.ppf(1 - alpha/2)

# Margin of error
m = Z_crit* pooled_variance

# Confidence Interval
CI = [(mu_con - mu_exp) - m, (mu_con - mu_exp) + m]

print("Test Statistics stat: ", T)
print("Z-critical: ", Z_crit)
print("P_value: ", p_value)
print("Confidence Interval of 2 sample Z-test for proportions: ", np.round(CI,2))

import matplotlib.pyplot as plt
z = np.arange(-3,3, 0.1)
plt.plot(z, norm.pdf(z), label = 'Standard Normal Distribution', color='purple', linewidth=2.5)
plt.fill_between(z[z>Z_crit], norm.pdf(z[z>Z_crit]), label="Right Rejection Region", color="y")
plt.fill_between(z[z<(-1)*Z_crit], norm.pdf(z[z<(-1)*Z_crit]), label="Left Rejection Region", color="y")
plt.title("Two Sample Z-test rejection region")
plt.legend()
plt.show()
```

## 8c. Chi Squared Test
If you want to test whether there is a statistically significant difference between the Control and Experiment groups' metric (for example conversions) and you don't really want to know the nature of this relationship (which one is better)

    - Metric should be in the form of a binary variable (conversion/no conversion, click/no click)
    - Data can be represented in the form of observed and theoretical
   
   
    - Chi Squared Expected Value is equal to the number of times each version of the product is viewed 
        multiplied by the probability of it leading to a conversion
    - Chi Squared is not a parametric test, its Std Error and CI cant be calculated the same way as T or Z Test

```python
import numpy as np
from scipy.stats import chi2

O = np.array([86, 83, 5810,3920])
T = np.array([105,65,5781, 3841])

# Squared_relative_distance

def calculate_D(O,T):
    D_sum = 0
    for i in range(len(O)):
        D_sum += (O[i] - T[i])**2/T[i]
    return(D_sum)

D = calculate_D(O,T)
p_value = chi2.sf(D, df = 1)


import matplotlib.pyplot as plt
# Step 1: pick a x-axis range like in case of z-test (-3,3,0.1)
d = np.arange(0,5,0.1)
# Step 2: drawing the initial pdf of chi-2 with df = 1 and x-axis d range we just created
plt.plot(d, chi2.pdf(d, df = 1), color = "purple")
# Step 3: filling in the rejection region
plt.fill_between(d[d>D], chi2.pdf(d[d>D], df = 1), color = "y")
# Step 4: adding title
plt.title("Two Sample Chi-2 Test rejection region")
# Step 5: showing the plt graph
plt.show()
```

### Standard Error and Confidence Interval for Non-Parametric Tests
To calculate Std Error and CI, we need to use resampling techniques like Bootstrapping and Bootstrap Quantile method

**Bootstrapping**: Takes original training sample and resamples from it by replacement, resulting in *B* different samples.
    The same observation can occur more than once in the bootstrap dataset.
    
    - Resample with replacement from the existing data of the two groups
    - Do this *B* times which means you will end up with *B* samples for two groups
    - Then calculate the sample means/medians for the control and experiment groups 
    - Can also calculate the difference in sample means for each each pair of control and experimental groups
    - Then if *B* is large, we can use the CLT ans assume the sampling distribution of difference of means follows normal distrib
    
   To obtain the p-value, we need to count the number of times out of *B* times, the difference in means is larger than 0
       
       - If p > the chosedn significance levem, then we can't reject the null. Hence there is not enough evidence 
           to state that there is a statistically significant difference in the control and experimental sample means
    
   
   
   To calculate the 95% Confidence Interval, one can use the Percentile Method, which uses the 2.5th and 97.5tth percentile of the bootstrap distribution of estimates as the lower and upper bounds of the interval.

# 9. Statistical Significance vs Practical Significance
After statistical significance is detected, the next step is to understand whether there is a *practical significance*.
This will help understand whether the detected difference in the performaces is large enough to justify the investment or it's too small and make a launch decision is not worth the investment.

To determine practical significance, is to use the Confidence Interval and compare the lower bound to the MDE (Min detectable effect, estimate of economic significance)
    - If CI lower bound > MDE, then you can state there is practical significance
    - Also look at width of CI, It shouldn't be too big since CI gives you the indication 
        that the precision of your results is small and the results will not be generalized to the entire population
        

# 10. A/B Test Quality
A/B Testing is one example of experimental design, and there are 3 factors needed 
to be satified to make solid conclusions about experiments:
    - Reliability / Replicability
    - Validity
    - Potency
    
    a. Reliability
        - Experimental results must be more than a one-off finding and be inherently reproducable and repeatable
    b. Validity
        - Encompasses the entire concept of your experiement and establishes whether the obtained results
            meet all the reqs of the randomized control trials or not
        - Internal Validity: Observed data and results obtained with it
            * Are the results valid and reliable?
            * Are the change in the dependent var only due to the independent var & not other factors?
            * Ex. omitted variable bias, reverse causality
        - External Validity: The degree to which the experimental results are generalizable to the entire population.
            * Can results be generalized to the entire population?
            * ex. biased sample, unrepresented sample, 
    

# 11. Common Problems and Pitfalls of A/B Test
A. Confounding Effects:

    - Need to ensure all other known possible factors that also have an impact on the dependent variable are held constant
    - Thus you need to control for as many unwanted/unequal (extraneous) factors as possible
        * Extraneous variables matter whwn they are associated with both indep & dep variables
    - Confounding effects threaten the internal Validity of your A/B experiment
    - Solutions to avoid this problem:
        * Control of confounding variables
        * Reliable instruments
        * Appropriate choice of dependent and independent variables
        * Generation of a random sample
       
B. Selection Bias:

    - One of the fundamental assumptions of A/B testing is that the sample is unbiased
    - To check for sample bias while knowing the true population distribution
        * Create *B* bootstrapped samples from the same & draw the distribution of sampling means
        * If the distribution is not centered around the true population mean, then it is biased


C. Systematic Bias: 

    - Related to the way one measure the impact of the treatment
    - Are you making errors when measuring it?
    - This type of error always affects measurements the same amount or by the same proportion 
        given that the reading is taken the same way each time - hence it is predictable
    - Unlike random error which affects the precision, systematic error affects the accuracy
    
    
D. Early Stopping or P-Hacking

    - A common mistake is to stop the experiment once you get a statistically sig result
    - While the significance level and all other model params are predetermined in the Power Analysis 
        and assume that the experiment will run until the minimum sample size is achieved
    - P-hacking or early stopping affects the Internal Validity of the results 
        and makes them biased so it also leads to false positives
    
    
E. Spillover or Network Effects

    - Applies to products where users in the experimental and control group are in the same communities 
        and influence each other, which leads to biased results
    - To detect this you can perform a **Stratified Sampling** (units in each subgroup are randomized)
        and then divide this into 2 groups
    - Then you can run an A/B test on one sample taking into account the clustered samples and the other one without
    - If there is a difference in treatment effects, then there is a network effect problem
    
    
F. Sample Ratio Mismatch

    - If the split between control and experimental looks suspicious, suggesting that the assignment process 
        is uneven, then you can perfom a Chi Square Test
    - This test will help check for Sample Ration mismatch
