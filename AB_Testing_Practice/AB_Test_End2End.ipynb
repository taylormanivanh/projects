# 1. What is AB Testing and when to use it?
The goal of A/B Testing is to show a new variation of a product to an experiemental group, and an existing version of the product to a control group. The goal is to track the metric during the test period and determine if there is a difference in performance/efficacy of the product - and what kind of difference is it.

Benefits of A/B Testing:

    - Quickly learn what works and what doesn't
    - Direct feedback from customers
    - Results are unbiased since testers are unaware
    
Drawbacks of A/B Testing:

    - Presenting different products/featurs/etc may be dangerous and casuse Change Aversion
    - Requires significant resources
    - Can lead to incorrect solutions if not conducted properly

# 2. Questions to ask before an A/B test

    - Can we assure truly randomized control and experiemental groups such that both samples are an unbiasedand 
        true representation of the true user population
    - Can we ensure teh intefrity of the treatment vs control eggects during the entire duration of the test?

# 3. Choosing Primary Metric for the A/B Test
It is important to chose the metric that will be used to measure the performance of the product. It will also be used to identigy whether there is a statistically significant difference between these two groups.

The choice of the *success metric* depends on the underlying hypotheis that is being tested with the A/B Test. We need to tie up the primary metric to the direcyt and higher level goals of the product. One way to test the accuracy of a metric you have chosen for your A/B Test could be to go back to the exact problem you want to solve.

*Metric Validity Question: If this chosen metric were to increase significantly while everything else stays constant, would we achieve our goal address the problem?*

Though you need to have a single primary metric for the A/B Test, we still need to keep an eye on the remaining metrics to make sure all the metrics are showing change and not only the target one. Having *multiple* metrics in the test will lead to false positives since you will identify many signigicant differences while there is no effect.

# 4. Common A/B Test Metrics
a. Click Through Rate: The percent of people who view the page (impressions) and then actually click on it
    
    CTR = (# total clicks * 100) / (# total clicks + # total views)



b. Click Through Probability: If a user has performed multiple clicks in a single session on the same item, CTP takes into account  
    duplicate clicks as a single click
    
    CTP = (# of people with at least 1 click * 100) / (# of unique visitors per page)
   
   
   
c. Conversion Rate: The proportion of sessions ending up with a transaction
   
    CR = # converted / (# converted + # not converted)

# 5. Stating the Hypothesis of the Test
A/B Test should always be based on a hypothesis that needs to be tested. The idea behind the hypothesis is to decide how to fix a potential issue in the product where a solution of these problems will influence the Key Performance Indicators (KPIs)

Don't merge multiple ideas into one hypothesis and also limit the variables introduced in the test so that you can understand their individual impact. 

# 6. Designing the A/B Test
   ### Step 1: Statistical Hypothesis
        a. Null:
            - probability of control = probability of experiment
            - population mean of control = population mean of experiment
        b. Alternate:
            - probability of control != probability of experiment
            - population mean of control != population mean of experiment
     
     
   ### Step 2: Power Analysis 
        Before running an A/B Test we need to determine the sample size of the control and
         experiemental groups and for how long we need to run the test. 
         This is known as the Power Analysis and it includes 3 steps:
             
         a. Determining the power of the test
             - The probability of correctly rejecting the null hypothesis when it is false.
             - Power = probability of not makeing a type 2 error
             - Common to pick 80% as the power of the test - 20% type 2 error which means we are fine with 
                 not detecting (failing to reject the null) a tratment effect while there is an effect
                 
         b. Determining the signigicance level of the test
             - Significance level = the probability of Type 1 error is the likelihood of rejecting the null,
                 hence detecting the treatment effect, while the null is true and there is no stat sig. impact.
                 (False pos rate)
             - Significance value of 5% indicates a 5% risk of concluding that there exists a stat sig. 
                 difference btwn the experimental and the control when there is no actual difference
                     
         c. Determining a Minimum Detectable Effect 
             - MDE = what is the substantive to the stat sig. that the business wants to see as a 
                 min impact of the new product?
             - What is the amount of change we aim to observe in the new product compared to 
                 the existing one to recommend to the business that the feature should be launched
             - MDE is a proxy that related to the smallest effect that would matter in practice for the business 
                 (usually set by stakeholders)


   ### Step 3: Calculating Minimum Sample Size
        a. Sample size must be determined using the:
            - power of the test
            - significance level
            - MDE
            - variances of the 2 normally distributed samples of equal size
            
        b. Case 1: Sample Size (N) Calculation with Binary Metric
            - CTR: user can click (success) or not (failure) and if the users' responses to the product
                can be defined as independent we can consider this *Bernoulli Trials* 
            - Bernoulli Trials: where the click event (success) occurs with probability p_con, p_exp (control/exp)
                - q_con = 1 - p_con
                - q_exp = 1 - p_con
            - The sample size needed to compare these two Binomial Proportions using a two sided test with prespecified
                sig level, power level, MID.
            - We need to use A/A testing (AB test except assigning the same treatment to both groups) 
                to obtain estimates.

         
         
        c. Case 2: Sample Size (N) Calculation with Continuous Metric
            - Primary performance metric that is in the form of an average (mean order amount) where we intend
                to compare the means of the contol & exp groups, then use the CLT and state the mean sampling
                distribution of both groups follow normal distribution
            - The same size needed to compare the means of 2 normally distributed samples using a 2 sided test with
                prespecified sig level, power, and MID
            - We can run an A/A test to obtain the sample variances _con, and _exp.

    
    
   ### Step 4: Determining A/B test duration
        a. Baseline duration time = N / # visitors per day
        b. Take into account business aspects (ex. Covid)

# 8. Analyzing A/B test results with Python
    Values to test whether there is a statistically significant differnce between control and experimental groups:
    - Choosing an appropriate statistical test
    - Calculating the test statistics
    - Calculating the p-value of test statistics
    - Reject or fail to reject the statistical hypothesis
    - Calculate the margin of error (external validity of the experiment)
    - Calculate confidence interval (external validity and practical significance of the experiment)
    
    a. Choosing the Appropriate Statistical Test
        i. Choose an appropriate stats test that is usually categorized by parametric or non-parametric, and depends on:
             - format of the primary metric
             - sample size (for CLT)
             - nature of the statistical hypothesis (show relationship btwn 2 groups exists or identigy type of relationship)
         ii. The most popular parametric tests in A/B Testing are:
             - 2 Sample T-Test: When N < 30, metric follows student-t distribution, and you want to identify
                whether there exists a relationship & the type of relationship btwn contol & exp
             - 2 Sample Z-Test: When N > 30, metric follows asymptotic Normal distribution and you want to identify
                whether there exists a relationship & the type of relationship btwn contol & exp
         iii. The most populat non-parametric tests in A/B Testing are:
             - Fisher's Exact Test: Small N, identigy and want to identify whether there exists a relationship
             - Chi-Squared Test: Large N, identify and want to identify whether there exists a relationship
             - Wilconxon Rank Sum/Mann Whitney U Test: Small or Large N, skewed sampling distributions, 
                 testing the difference in medians between control and experiemental groups.
         

## 8a. 2-Sample T-Test
If you want to test whether there is a statistically significant difference between the control and experiemental groups' metrics
    that are in the form of averages, metric follows student-t distribution and N < 30

```python
import numpy as np
from scipy.stats import t

np.random.seed(18)

N_con = 20
df_con = N_con - 1 # degrees of freedom of Control
N_exp = 20
df_exp = N_exp - 1 # degrees of fredom of Experiment

# Significance Level
alpha = 0.05

# Data of control group with t-distribution
X_con = np.random.standard_t(df_con, N_con)
# Data of experimental group with t-distribution
X_exp = np.random.standard_t(df_exp, N_exp)

# Mean of Control
mu_con = np.mean(X_con)
# Mean of Experimental
mu_exp = np.mean(X_exp)

# Variance of Control
sigma_sqr_con = np.var(X_con)
# Variance of Experimental
sigma_sqr_exp = np.var(X_exp)

# Pooled Variance
pooled_var_t_test = ((N_con - 1) * sigma_sqr_con + (N_exp - 1) * sigma_sqr_exp) / (N_con + N_exp - 2)*(1/N_con + 1/N_exp)

# Standard Error
se = np.sqrt(pooled_var_t_test)

# Test Statistics
T = (mu_con - mu_exp) / se

# Critical value for 2 sided sample t test
t_crit = t.ppf(1-alpha/2, N_con + N_exp - 2)

# P value of the 2 sided t testm using t-distribution and its symmetric property
p_value = t.sf(T, N_con + N_exp - 2) * 2

# Margin of Error
margin_error  = t_crit * se

# Confidence Interval
ci = [(mu_con - mu_exp) - margin_error, (mu_con - mu_exp) + margin_error]

print("T-score: ", T)
print("T-critical: ", t_crit)
print("P_value: ", p_value)
print("Confidence Interval of 2 sample Z-test: ", np.round(ci,2))
```
